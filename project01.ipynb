 Exercise 1: Column information

```python
import pandas as pd

df = pd.read_csv("proj1_ex01.csv")
df
```

```python
toJson = {
    'name': df.columns,
    'missing': df.isna().sum()/len(df.index),
    'type': df.dtypes
}

toJsonDf = pd.DataFrame(data = toJson)
result = toJsonDf.to_json("proj1_ex01_fields.json", orient='records', default_handler=str)
```

Exercise 2: Value statistics

```python
import json
```

```python
stats = df.describe()
stats
```

```python
import numpy as np
stats2 = df.describe(exclude=[np.number])
stats2
```

```python
stats_json_dict = stats.to_json()
stats_json = json.loads(stats_json_dict)
stats_json
```

```python
stats2_json_dict = stats2.to_json()
stats2_json = json.loads(stats2_json_dict)
stats2_json
```

```python
stats_json.update(stats2_json)
stats_json
```

```python
with open('proj1_ex02_stats.json', 'w') as file:
    json.dump(stats_json, file)
```

Exercise 3: Column names

```python
names = df.columns.to_list()
names
```

```python
import re

for idx, name in enumerate(names):
    names[idx] = names[idx].lower()
    names[idx] = names[idx].replace(" ", "_")
    names[idx] = re.sub('[^0-9a-zA-Z]+', '_', names[idx])

names
```

```python
df = df.set_axis(names, axis='columns')
df
```

```python
df.to_csv("proj1_ex03_columns.csv", index=False)
```
